```python
#Task 5
class StudentDatabase:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = {}
        self.course_name=[]
    def calculateGPA(self, courses, semester):
      self.semester=semester
      course1=0
      course2={}
      course_name=[]
      for i in range(len(courses)):
        course=courses[i].split(":")
        course_name+=[course[0]]
        course1+=float(course[1])*3
      course_sum=course1/(len(course_name)*3)
      course_sum=round(course_sum,2)
      course2[tuple(course_name)]=course_sum
      self.grades[self.semester]=(course2)
      self.grades.update(self.grades)

    def printDetails(self):
      print(f"Name: {self.name}")
      print(f"ID: {self.student_id}")
      for key in self.grades:
        print(f"Courses taken in {key}:")
        for j in self.grades[key]:
          for k in list(j):
            print(k)
          print(f"GPA: {self.grades[key][j]}")





s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```

```python
#Task 6
class Store:
  def __init__(self,balance):
        self.balance = balance
        self.items = {}
        self.total_items=""
        print("New branch created!")

  def viewAllItems(self):
      if self.items=={}:
        print("There are no itmes in your inventory")
      else:
          for key in self.items:
            self.total_items+=key
            if key!=list(self.items.keys())[-1]:
              self.total_items+=","
          print(f"All Items: {self.total_items}")
  def viewAllItemDetails(self):
    print(self.items)
  def add_item(self,items):
    self.items[items[0]]={'stock':items[1],'buying_price':items[2],'selling_price':items[3]}
    self.items.update(self.items)
    print(f"Item added: {items[0]}")
    self.balance=self.balance-(items[2]*items[1])
  def sell_item(self,x,y):
    if y>self.items[x]['stock']:
      print(f"Sorry! ChaCha Noodles is not available at your desired quantity.Currently we have: {self.items[x]['stock']} Current Balance: {self.balance}")
    else:
      sell=y*self.items[x]['selling_price']
      self.balance+=sell
      self.items[x]['stock']-=y


  def restock_item(self,a,b):
    self.items[a]['stock']+=b
    print(f'Restocked item: {a},',"Current Stock:",self.items[a]['stock'])
    self.balance-=self.items[a]['buying_price']*b
print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```
